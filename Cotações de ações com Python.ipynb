### Objetivo: buscar a cotação de dois ativos, e fazer a comparação de ambos em um período de tempo:

- 1º importando bibliotecas importantes:

```python
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt
```

- 2º definindo o período que se quer a cotação:

```python
inicio = '01/01/2020'
fim = '11/01/2022'
```

- 3º definindo quais ativos você quer ver a cotação:

    - Pode escolher digitar o ticker, ou então ler um banco de dados com varios tickers.

```python
ticker = 'BOVA11.SA','PETR4.SA'
```

- 4º criando um dataframe que irá conter a leitura dos dados retirados da biblioteca yf, contendo os parâmetros estipulados:

```python
cotacao_b3 = web.DataReader(ticker, data_source = 'yahoo', start = inicio , end= fim)['Adj Close']
display(cotacao_b3)
```

- 5º plotagem de um gráfico de acordo com dados coletados, dimensões do gráfico alteradas para melhor visualização;

```python
cotacao_b3.plot(figsize=(15,10));
```

- 6º selecionando os dados do dataframe da primeira linha até o final -> método .iloc
---
- 7º normalizando os dados para análise de retorno acumulado, partindo de um mesmo ponto.

```python
cotacao_b3.iloc[0]
cotacao_normalizada = cotacao_b3/cotacao_b3.iloc[0]
cotacao_normalizada.plot(figsize=(15,10));
```
